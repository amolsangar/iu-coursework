    Answer - 

    To determine if Gv is connected or not, check if there's any node in G connected to any node in Gv by an edge.

    If there's an edge, then Gv is connected to G otherwise its a disconnected graph.

    To do this, we can use traversal algorithms like DFS or BFS to see if they visit all nodes or not. To run it in O(m+n) time, we can maintain an adjacency list that stores edges from every node. Thus, we are checking the whole graph only once in O(m+n) time to determine the connectivity.

```python
class Graph:
    def __init__(self, edges, n):

        self.adjList = [[] for _ in range(n)]

        # add edges to the directed graph
        for (src, dest) in edges:
            self.adjList[src].append(dest)

def DFS(graph, v, visited):
    visited[v] = True

    for u in graph.adjList[v]:
        if not visited[u]:
            DFS(graph, u, visited)

# Function to check graph connectivity
def isConnected(graph,n):
    # n is no of nodes
    visited = [False] * n
    v = 0
    
    DFS(graph, v, visited)
    
    for v in visited:
        if not v:
            return False
    
    return True
```

### Problem 7 [5 pt.]

What is the time complexity of the function calc in big-Oh? What is the return value k in terms of big-Oh notion? For full credit please show your work step by step.

Answer - 

Time complexity of calc() is dependent upon 2 loops. First loop takes O(n/2) time (n/2 -> n). The second loop takes O(log.n) time since j is multiplied by 2 on every iteration and thus the overall iterations keeps getting reduced by half.

Thus overall time complexity is O(n * log.n) because of the dominant terms

The return value of k is n/4 * O(n * log.n)

```python
def calc(n):
    k = 0
    i = n/2
    while(i<=n):     # Time complexity here is O(n/2)
        i = i + 1
        j = 2
        while(j<=n):    # Time complexity here is O(log.n)
            j = j * 2
            k = k + n/2    # Result = n/4 * O(n * log.n)
    return k
```

```python
# Define an undirected graph
class UndirectedGraph:
    def __init__(self, n):
        self.num_nodes = n
        self.nodes = [set() for i in range(n)]
        self.edges = []
    
    # edge node1 <--> node2 (undirected)
    def insert(self, node1, node2, weight=1.0):
        self.nodes[node1].add(node2)
        self.nodes[node2].add(node1)
        self.edges.append(Edge(node1, node2, weight))
        self.edges.append(Edge(node2, node1, weight))
```

```python
# Union Find data structure is provided to support Kruskal
class UnionFind:
    def __init__(self, num_nodes):
        # Initially position[i] = i
        self.position = [i for i in range(num_nodes)]
        
    # Return the cluster index
    def find(self, node):
        if self.position[node] == node:
            return node
        else:
            self.position[node] = self.find(self.position[node])
            return self.position[node]
    
    def union(self, node1, node2):
        a = self.find(node1)
        b = self.find(node2)
        # no need to union
        if a == b:
            return
        # union is needed
        else:
            if a < b:
                self.position[b] = a
            else:
                self.position[a] = b
```

```python
# Implement Kruskal

# returned MST is a list of class Edge
def kruskal(graph):
    MST = []
    uf = UnionFind(graph.num_nodes)

    sorted_edge_list = MergeSort(graph.edges)
    
    n = len(sorted_edge_list)
    
    while(len(MST) != n-1 and len(sorted_edge_list) > 0):
        edge = sorted_edge_list.pop(0)
        edge_a = edge.node1
        edge_b = edge.node2

        a = uf.find(edge_a)
        b = uf.find(edge_b)

        if (a != b):
            MST.append(edge)
            uf.union(edge_a, edge_b)
    
    return MST
```

A modified knapsack problem is that for a given set of positive integers {a1, a2, ..., an}, a knap- sack of size s, find a subset A of {a1, a2, ..., an} such that the sum of elements in A is the largest, but at most s. Write a program using Python to solve this problem using a top down dynamic programming method.

Test your program for the following modified knapsack problem:
Input list: {5, 23, 27, 37, 48, 51, 63, 67, 71, 75, 70, 83, 889, 91, 101, 112, 121, 132, 137, 141, 143, 147, 153, 159, 171, 181, 190, 191} with knapsack size s = 762. <br>Print out a subset along with the sum of its elements so that the sum has the closest distance to 762.

```python
import copy
```

```python
def trace_subset(subset, input_list, max_size):
    knapsack_subset = set()
    
    size = copy.deepcopy(max_size)
    
    for i in range(len(input_list)):
        key = (i, size)
        if key in subset.keys():
            edge_weight = subset.get(key)[1]
            if edge_weight:
                size -= input_list[i]
                print(f"Weight {input_list[i]} added from index {i}")
                knapsack_subset.add(input_list[i])

    return knapsack_subset 
```

```python
def knapsack_subset_recursive(subset, input_list, size, current_index):
    if size <= 0 or current_index >= len(input_list):
        return 0

    key = (current_index, size)
    if (key in subset.keys()):
        return subset.get(key)[0]

    # Left Branch - Value included 
    include = 0
    current_element = input_list[current_index]
    if (current_element <= size):
        include = current_element + knapsack_subset_recursive(subset, input_list, (size - current_element), current_index+1)

    # Right Branch - Value excluded 
    exclude = knapsack_subset_recursion(subset, input_list, size, 1 + current_index)

    edge_weight = 0
    if (include > exclude):
        edge_weight = 1
    subset[(current_index, size)] = (max(include, exclude), edge_weight)

    return max(include, exclude)
```

```python
def knapsack_subset(input_list, max_size):  
    subset = {}
    current_index = 0
    knapsack_value = knapsack_subset_recursion(subset, input_list, max_size, current_index)
    knapsack_subset = trace_subset(subset, input_list, max_size)
    
    return (knapsack_value, knapsack_subset)
```

```python
A = [5, 23, 27, 37, 48, 51, 63, 67, 71, 75, 70, 83, 889, 91, 101, 112, 121, 132, 137, 141, 143, 147, 153, 159, 171, 181, 190, 191]
s = 762

print("Knapsack Size:", s)

print()
(knapsack_value, knapsack_subset)= knapsack_subset(A, s)
print("\nOutput: ")
print("Knapsack value: {}".format(knapsack_value))
print("Knapsack subset: {}".format(knapsack_subset))
```

### Explanation: 

To find the closest value to the knapsack using top down approach, we are branching the input elements where values are subtracted from the knapsack value in the left branch and kept as it is in the right branch. We recursively do this operation until we run out of elements. To make this recursive approach dynamic, max elements are picked up from the memory saved in the form of a dictionary. Once the max knapsack value is calculated, dictionary lookup is performed to get all index values that forms the final knapsack subset.
